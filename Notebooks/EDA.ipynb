## Feature Engineering
Separação e extração de variáveis textuais e temporais  (Localizacao, Data_Lancamento e Detalhes_da_Missao) em novas colunas atômicas, estruturadas e tipadas (Pais, Ano, Modelo_Foguete, etc.) para otimização da análise, seguida do descarte de dados redundantes.


## Decomposição dos Detalhes da Missão (Modelo_Foguete e Carga_Util)
O que são:  
- Modelo_Foguete: É o veículo lançador, a máquina física responsável por vencer a gravidade da Terra (ex: Falcon 9, Ariane 5, Soyuz).

- Carga_Util (Payload): É o objeto ou tripulação que está sendo transportado para o espaço (ex: um satélite de internet da Starlink, o robô Perseverance enviado para Marte, ou uma cápsula com astronautas).

Por que separamos:
Na base bruta, essas duas informações vinham coladas na mesma linha de texto, separadas por uma barra vertical (|). Se mantivéssemos os dados juntos, seria impossível responder a perguntas cruciais de negócio de forma isolada, como: "Qual é o modelo de foguete mais utilizado na história?" ou "Quantos satélites de pesquisa foram enviados?". A separação transforma um bloco de texto confuso em duas métricas limpas, permitindo filtros exatos no Dashboard e análises estatísticas individuais para cada componente da missão.

```
# Tratamento da Coluna de Localização
df['Pais'] = df['Localizacao'].str.split(',').str[-1].str.strip()
df['Estado_Regiao'] = df['Localizacao'].str.split(',').str[-2].str.strip()

# Tratamento da Coluna de Data e Hora
df['Data_Auxiliar'] = pd.to_datetime(df['Data_Lancamento'].str.replace('UTC', ''), errors='coerce')

df['Ano'] = df['Data_Auxiliar'].dt.year
df['Mes'] = df['Data_Auxiliar'].dt.month
df['Dia'] = df['Data_Auxiliar'].dt.day
df['Hora'] = df['Data_Auxiliar'].dt.hour

# Garantindo o preenchimento do ano caso a conversão de data falhe em linhas antigas
df['Ano'] = df['Ano'].fillna(df['Data_Lancamento'].str.extract(r'([1-2][0-9]{3})')[0].astype(float))

# Corrigindo a formatação para números inteiros puros
colunas_tempo = ['Ano', 'Mes', 'Dia', 'Hora']
for col in colunas_tempo:
    df[col] = df[col].astype('Int64')

# Removendo a coluna auxiliar
df = df.drop(columns=['Data_Auxiliar'])

# Tratamento dos Detalhes da Missão
df[['Modelo_Foguete', 'Carga_Util']] = df['Detalhes_da_Missao'].str.split('|', n=1, expand=True)

df['Modelo_Foguete'] = df['Modelo_Foguete'].str.strip()
df['Carga_Util'] = df['Carga_Util'].str.strip()

# Apagando a coluna original já que as informações foram extraídas
df = df.drop(columns=['Detalhes_da_Missao','Data_Lancamento', 'Localizacao'])
```

## Estratégia de Tratamento de Dados Faltantes (Missing Values)
Ao verificar os valores nulos da nossa base, identificamos dois cenários distintos que exigem tratamentos diferentes:

1. Dados Temporais (Mês, Dia e Hora)
Temos 126 registros vazios. Tratam-se de missões históricas antigas (décadas de 50, 60 e 70) onde essas informações não foram documentadas com precisão na época.

Decisão: Manter esses valores em branco (Nulos). Não faremos preenchimentos para preservar a veracidade histórica dos dados.

2. Dados Financeiros (Custo da Missão)
Identificamos 3.360 valores vazios (cerca de 78% da base). Remover essas linhas destruiria nosso dataset. Portanto, traçamos uma estratégia estatística para preencher essas lacunas sem distorcer a realidade do mercado.

 **O Erro da Primeira Tentativa e a Mudança de Planejamento
Inicialmente, tentamos aplicar uma abordagem padrão combinando Imputação Condicional pela Empresa (Plano A) com um Fallback pela Mediana Global (Plano B). No entanto, a validação dos dados revelou duas grandes distorções (outliers):**

O caso RVSN USSR (União Soviética): Como a agência soviética possuía um foguete histórico (Energia) com custo registrado absurdamente alto (5.000 milhões), a mediana da empresa ficou travada nesse teto. A imputação simples fez com que todos os seus lançamentos comuns e antigos herdassem erroneamente esse custo bilionário, inflando drasticamente a média histórica.

O "Efeito 450.0": A mediana global pura da base estava fixada em 450 milhões devido ao peso de grandes agências modernas. Isso fez com que empresas menores que não tinham nenhum dado financeiro (como a AEB - Agência Espacial Brasileira) recebessem valores irreais por seus lançamentos.

## Novo Planejamento Adotado (Imputação Refinada por Camadas)
Para corrigir essas distorções e blindar nossa análise, mudamos o planejamento para uma estratégia hierárquica muito mais precisa e contextualizada:

Camada de Segurança (Tratamento de Outliers): Identificamos e isolamos manualmente o impacto da RVSN USSR, aplicando um teto de custo realista (150 milhões) para suas missões comuns que estavam sem registro.

- Imputação Condicional por Empresa (Plano A): Mantivemos o preenchimento dos custos vazios usando a mediana da própria empresa responsável, respeitando o perfil operacional de gastos de cada agência (como a diferença entre os altos custos da NASA e os custos comerciais da SpaceX).

- Imputação Condicional por Ano/Década (Plano B - Novo Fallback): Em vez de usar uma mediana global genérica para empresas sem dados, o código agora busca a mediana das missões ocorridas naquele mesmo ano. Isso respeita a evolução da economia e da tecnologia (foguetes antigos eram mais baratos que foguetes modernos).

- Mediana Global Ajustada (Plano C): Como último recurso para casos raros e isolados, aplicamos a mediana global, porém filtrando e ignorando os grandes outliers numéricos da base para manter a integridade dos dados.

Dessa forma, os dados pós-imputação passaram com sucesso na validação, apresentando médias perfeitamente coerentes com a realidade histórica e econômica.

```
# Certificando que a coluna é numérica
df['Custo_da_Missao'] = df['Custo_da_Missao'].astype(str).str.replace(',', '', regex=False).str.strip()
df['Custo_da_Missao'] = pd.to_numeric(df['Custo_da_Missao'], errors='coerce')

# [Ajuste de Segurança] A RVSN USSR tem um outlier de 5000 que distorce a história.
# Vamos limitar o custo máximo dela à mediana das outras agências da mesma época se for nulo.
df.loc[(df['Nome_da_Empresa'] == 'RVSN USSR') & (df['Custo_da_Missao'].isna()), 'Custo_da_Missao'] = 150.0 # Custo estimado realista de um foguete soviético padrão

# Imputação por Empresa (Plano A)
mediana_por_empresa = df.groupby('Nome_da_Empresa')['Custo_da_Missao'].transform('median')
df['Custo_da_Missao'] = df['Custo_da_Missao'].fillna(mediana_por_empresa)

# Imputação por Ano/Década (Plano B - Muito mais preciso que a global pura)
mediana_por_ano = df.groupby('Ano')['Custo_da_Missao'].transform('median')
df['Custo_da_Missao'] = df['Custo_da_Missao'].fillna(mediana_por_ano)

# Se ainda sobrar algum nulo (empresas isoladas em anos sem dados), usamos a mediana global baixa
mediana_geral_ajustada = df[df['Custo_da_Missao'] < 1000]['Custo_da_Missao'].median() # Ignora o outlier de 5000 no cálculo geral
df['Custo_da_Missao'] = df['Custo_da_Missao'].fillna(mediana_geral_ajustada)

print(df.groupby('Nome_da_Empresa')['Custo_da_Missao'].mean().sort_values(ascending=False).head(5))
```
